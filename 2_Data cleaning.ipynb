```python
# Imports
import numpy as np
import pandas as pd

# Importing data

df = pd.read_csv('data/eda.csv', delimiter=",")

df.head(10)
```

## Improving columns

```python
# Dropping house_id column
df = df.drop(['house_id'], axis=1)
df.head()
```

```python
#Checking for missing values overall
df.isna().sum()
```

```python
# Renaming first column from id --> house id,
# renaming id.1 column to transaction_id

df = df.rename(columns={"id": "house_id", "id.1": "transaction_id"})
df.head()
```

## Cleaning cells

```python
# Transforming datatype for uncomplicated columns

df = df.astype({'bedrooms': 'int32',
                'bathrooms': 'int32',
                'sqft_living': 'int32',
                'sqft_lot': 'int32',
                'floors': 'int32',
                'sqft_living15':'int32',
                'sqft_lot15':'int32'})
df.head()
```

```python
# Checking for duplicates

df.duplicated().value_counts()
```

```python
# fixing Waterfront
df['waterfront'].unique()

# assuming 'Nan' values are not located at a waterfront --> filling with 0
df = df.fillna({'waterfront':0})
df = df.astype({'waterfront': 'int32'})

df['waterfront'].unique()
```

```python
# fixing view
df['view'].unique()
#df.query("`view` not in [0, 1, 2, 3, 4]")

# setting 0 for NaN values
df = df.fillna({'view': 0})
df = df.astype({'view': 'int32'})

df['view'].unique()

```

# fixing year renovated



```python
df["yr_renovated"].unique()

df['yr_renovated'] = df.yr_renovated.apply(lambda x: x/10)
df = df.fillna({'yr_renovated': 0})
df = df.astype({'yr_renovated': 'int32'})

df["yr_renovated"].unique()
```

## fixing sqft_basement

```python
df['sqft_basement'].unique()


df = df.fillna({'sqft_basement': 0})
df = df.astype({'sqft_basement': 'int32'})

df["sqft_basement"].unique()
```

```python
df.isna().sum()
```

```python
df.to_csv('data/eda_cleaned.csv')
```
